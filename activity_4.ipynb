```python
import pandas as pd

data = {
    "Department": ["Operations", "Operations", "Operations",
                   "Marketing", "Marketing",
                   "IT", "IT"],
    "Category": ["Salaries", "Equipment", "Training",
                 "Advertising", "Events",
                 "Software", "Hardware"],
    "Amount": [50000, 20000, 10000,
               30000, 15000,
               25000, 35000]
}

df = pd.DataFrame(data)
df
```

```python
!pip install plotly -q

import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
fig = px.sunburst(
    df,
    path=["Department", "Category"],
    values="Amount",
    title="Company Budget Distribution"
)

fig.update_layout(
    title_x=0.5,
    width=800,
    height=600
)

fig.show()
```

```python
departments = list(df["Department"].unique())
categories = list(df["Category"].unique())

labels = departments + categories

source = []
target = []
value = []

for _, row in df.iterrows():
    source.append(labels.index(row["Department"]))
    target.append(labels.index(row["Category"]))
    value.append(row["Amount"])

fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels
    ),
    link=dict(
        source=source,
        target=target,
        value=value
    )
)])

fig.update_layout(
    title_text="Flow of Company Budget",
    title_x=0.5,
    font_size=12,
    width=900,
    height=600
)

fig.show()
```

```python
fig.update_layout(
    template="plotly_dark",
    title_x=0.5
)
```

```python
new_row = pd.DataFrame({
    "Department":["IT"],
    "Category":["Cyber Security"],
    "Amount":[20000]
})

df = pd.concat([df, new_row], ignore_index=True)
```

```python
fig = px.sunburst(
    df,
    path=["Department", "Category"],
    values="Amount"
)

fig.update_traces(
    hovertemplate=
    "<b>%{label}</b><br>" +
    "Amount: %{value}<br>" +
    "Percentage: %{percentParent:.2%}<extra></extra>"
)

fig.show()
```

```python
fig = go.Figure(data=[go.Sankey(
    node=dict(label=labels),
    link=dict(
        source=source,
        target=target,
        value=value,
        customdata=value,
        hovertemplate=
        'Flow: %{value}<br>' +
        'Percentage: %{value}<extra></extra>'
    )
)])
```
